```python
y= numpy.loadtxt('samples/1.raw', skiprows=1, usecols=1, delimiter=",")
for i in range(len(y)):
    dx = 0.1
    dy = diff(y)/dx
#the code above finds the difference in 'y' divided by the difference in 'x' (estimated slope) for every interval of the data I imported using numpy.loadtxt
#this is so that I can identify where the slop begins to be negative

#In the code below, I break off the slope calculation (from above) where it starts to be negative.
#From the printed statement, I can identify where the calculation breaks off(row "12")
#bp stands for "breakpoint" 
bp = -1
for i in dy:
    bp += 1
    if i < 0:
        break 
print(bp)
```

```python
#put previous code into a function 
def young_mods(data):
    for i in range(len(data)):
        dx = 0.1
        dy = diff(data)/dx
    bp = -1
    for i in dy:
        bp += 1
        if i < 0:
            break 
#below is youngs modulus formula for data until breakpoint(bp)
    E = data[12]/(a[:,0][12]) 
    return(E[1])
```
